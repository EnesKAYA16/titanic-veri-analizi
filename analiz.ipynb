```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=["Cabin", "Ticket", "Name", "PassengerId"])
df["Age"] = df["Age"].fillna(df["Age"].median())
df = df.dropna()
df.info()
```

```python
sns.countplot(data=df, x="Survived", hue="Sex")
plt.title("Cinsiyete Göre Hayatta Kalma")
plt.xticks([0, 1], ["Öldü", "Hayatta Kaldı"])
plt.show()
```

```python
sns.countplot(data=df, x="Pclass", hue="Survived")
plt.title("Bilet Sınıfına Göre Hayatta Kalma")
plt.xlabel("Sınıf (1=Birinci, 3=Üçüncü)")
plt.legend(["Öldü", "Hayatta Kaldı"])
plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", bins=30)
plt.title("Yaşa Göre Hayatta Kalma")
plt.legend(["Öldü", "Hayatta Kaldı"])
plt.show()
```
